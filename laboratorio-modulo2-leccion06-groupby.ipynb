#### Ejercicio 3: ¿Cuántos títulos hay por cada país? Ordena los resultados de mayor a menor

```python
# Creamos el groupby agrupando por países y contando el número de títulos por país.
titulos_pais = netflix_final.groupby("country")["Title"].count()
titulos_pais = titulos_pais.reset_index()
# Ordenamos la cuenta de los títulos de forma ascendente
titulos_pais.sort_values("Title", ascending = False)
```

Los países con mayor número de títulos son Estados Unidos, India, Reino Unido y Japón. Los que menos Corea del Sur, Alemania del Este, Colimbia y Venezuela.

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
# Comprobamos el tipo de datos que tiene la columna date_added. Vemos que contiene elementos de tipo objeto, y nosotros queremos tenerlo en tipo fecha.
netflix_final["date_added"].info()
```

```python
# Nos creamos una nueva columna que contenga solo el año en el que se añadió el título. 
netflix_final["año_añadido"] = pd.to_datetime(netflix_final["date_added"], format = "mixed").dt.strftime("%Y")
# Le hemos tenido que indicar que el formato de las fechas que contiene la columna no son todos del mismo tipo, sino que son diferentes. Por eso
# usamos el método stftime el cual me cambia las fechas de formato str a datetime de una columna concreta.
netflix_final.head(1)
```

```python
netflix_final["año_añadido"].info()
```

```python
titulos_por_año = netflix_final.groupby("año_añadido")["Title"].count().reset_index()
titulos_por_año.sort_values("Title", ascending = False)
```

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
# Agrupamos los títulos por país y fecha de lanzamiento, y realizamos una cuenta del número de título lanzados por año por país.
titulos_pais_2020 = netflix_final.groupby(["country", "release_year"])[["Title"]].count().reset_index()
# Aplicamos un filtro para que se quede solo con los títulos lanzados en 2020, los ordenamos de mayor a menor, y no quedamos con el primer elemento, que es el mayor.
titulos_pais_2020[titulos_pais_2020["release_year"] == 2020].sort_values("Title", ascending = False).head(1)

```

El país con más títulos lanzados en 2020 fue Estados Unidos.

```python
titulos_pais_2020[titulos_pais_2020["release_year"] == 2020].sort_values("Title", ascending = False).idxmax()
# Usando la función idxmax nos devuelve la posición del elemento de mayor valor. 
```

#### Ejercicio 7:¿Cuál es el promedio de puntuación en IMDb de los títulos agrupados por país?

```python
# Creamos un nuevo df que contiene el promedio de IMDB Score por país.
puntuacion_pais_titulos = netflix_final.groupby("country")["IMDB Score"].mean().reset_index().round(2)
puntuacion_pais_titulos[puntuacion_pais_titulos["IMDB Score"].notnull()]
# El df contiene valores nulos, pero de momento no nos interesa verlos, ya que estos no tienen puntuación, por lo que no aportan información
# sobre la puntuación media por país, sino sobre los países en los que no hay IMDB Score. Estos países se corresponderán con los títulos
# que son series de televisión o películas no originales de netflix.
```

#### Ejercicio 8: ¿Cuántos títulos fueron lanzados en cada año agrupados por tipo ('Movie' o 'TV Show')?
